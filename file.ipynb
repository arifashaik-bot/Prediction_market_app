```python
import pandas as pd
import os

# Get all parquet files in the current folder
files = [f for f in os.listdir() if f.endswith(".parquet")]
print("Found files:", files)

# Load all files into a list of dataframes
dfs = [pd.read_parquet(f) for f in files]

# Combine all into one dataframe
df = pd.concat(dfs, ignore_index=True)

# Show first 5 rows
df.head()
```

```python
# Show number of rows and columns
print("Total rows:", len(df))
print("Columns:", df.columns.tolist())
```

```python
# Show all column names in your dataset
df.columns.tolist()
```

```python
import matplotlib.pyplot as plt

# Use 'market_ticker' instead of 'market_name'
top_markets = df['market_ticker'].value_counts().head(10)

# Plot
top_markets.plot(kind='bar', title="Top 10 Prediction Markets")
plt.xlabel("Market Ticker")
plt.ylabel("Number of Trades")
plt.show()
```

# Conclusion:
The analysis of the Prediction Market dataset shows which markets are the most active and influential. By looking at the top 10 markets by trade count, we can see where public interest is highest. Examining YES and NO prices, as well as trade volumes over time, reveals trends in confidence and shifts in opinion. Overall, this data demonstrates how prediction markets reflect popular sentiment and can be used to identify the markets that have the greatest impact on public perception and decision-making.
